```python
import matplotlib.pyplot as plt
import pandas as pd
from pylab import rcParams
import seaborn as sns
rcParams['figure.figsize'] = 13,6
df = pd.read_csv('2015.csv')
data = df.groupby(['Region']).median().sort_values(by = 'Happiness Score', ascending = False)
data_1 = data[['Happiness Score']]
data_1.plot(kind='bar', 
                                                                  title='распределение уровня счастья по регионам',
                                                                  xlabel='регион',
                                                                  ylabel='уровень счастья')
# по этой диаграмме можем сделать вывод по среднему(на самом деле я взял медиану, так как я считаю здесь медиана уместнее)
# уровню счатья по регионам, на глаз оценить разницу между 'худшим' и 'лучшим' регионом
```

```python
ax = df.plot(kind='scatter', 
        x='Freedom', 
        y='Happiness Score', 
        s=df['Health (Life Expectancy)']*100,
        title='Correlation between Happiness Score and Freedom')
ax.grid()
#по этой диаграмме можно заметить как зависит уровень счатья от свободы, в среднем уровень счатья растет вместе со свободой
# можно отметить, что высокого уровня счатья с низкой свободы нет, а вот высокая свобода не гарантирует высокий уровень счастья
```

```python
ax = df[['Happiness Score']].plot(kind='hist', 
                     bins=10,
                     title='Happiness Score distribution',
                     xlabel='Happiness Score',
                     xticks=range(2, 8))
# по этой гистограмме видно распределение стран по уровням счатья, стран с очень низким уровнем счатья довольно мало 
# большинство стран находятся между 4 и 6 условными единицами, интересно отметить, что уровень счастья в примерно 5,5
# встречается гоаздо реже чем 5 или 6.
```

```python
data = df[['Economy (GDP per Capita)'] + ['Happiness Score']].sort_values(by = ['Economy (GDP per Capita)'], ascending = False)
data.set_index(['Economy (GDP per Capita)'], inplace = True)
ax = data.plot(ylabel ='Happiness Score', title = 'correlation between Economy and Hapiness Score' )
ax.grid()
# По графику видно, что с ростом економики уровень счатья растет, хотя это явно далеко не единственный фактор, интересно 
# что при высоких показателях экономики уровень счастья начинает вести себя более стабильно, хотя это может быть связано
# с тем, что высокая экономика у маленького количества стран
```
